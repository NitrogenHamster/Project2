```python
data_dropped = data_df.drop(columns = ["Url","id","bodytype","drivingconfiguration","Odometer", "Location"])
data_dropped
```

```python
data_dropped['brand'].unique()
```

```python
data_dropped.dtypes
```

```python
data_dropped = data_dropped[data_dropped.Price != "N/A"]
# data_df.dropna(subset=['Price'])
data_type = data_dropped.astype({'Price': 'float'},{'year':'int'})
```

```python
data_all = data_type.groupby(['year']).mean()
data_all
```

```python
data_brand = data_type.groupby(['brand','year']).mean()
data_brand.Price = data_brand.Price.round()
data_brand.head(50)
```

```python
data_model = data_type.groupby(['brand','model','year']).mean()
data_model.Price = data_model.Price.round()
data_model
# print(data_model["brand"])
```

```python
data_dic = data_brand.to_dict('series')
data_dic
```

```python
data_color_filtered = data_type[data_type.Color != "N/A"]
data_color = data_color_filtered.groupby(['year','Color']).count()
# data_color = data_color[data_color.Color != "N/A"]
# data_brand.Price = data_brand.Price.round()
data_color.head(50)
```

```python
data_color_all = data_color_filtered.groupby(['Color']).count()
data_color_all
```
