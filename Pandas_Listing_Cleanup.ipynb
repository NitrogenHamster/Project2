```python
import json
import pandas as pd
import numpy as np
```

```python
listings = pd.read_json("Listing_large_new.json")
listings
```

```python
# Drop unwanted columns 
listings = listings.drop(columns = ['id'])

# Edit headings
listings = listings.rename(columns={"Url":"Web Link","year":"Year","brand":"Manufacturer","model":"Model","bodytype":"Body Type","drivingconfiguration":"Drive Configuration"})
listings.head()
```

```python
listings.dtypes
```

```python
# Drop unwanted columns 
listings = listings.drop(columns = ["Body Type","Drive Configuration","Color","Location","Web Link"])
listings.head()
```

```python
# REMOVE NA FROM YEAR 
listings = listings[listings.Year !="N/A"]
listings.head()
```

```python
listings = listings[listings.Price !="N/A"]
listings.head()
```

```python
listings = listings.sort_values("Price", ascending = False)
listings
```

```python
# Reorganizing the columns using double brackets
listings = listings[["Year","Manufacturer","Model","Body Type","Drive Configuration","Odometer","Price","Color","Location","Web Link"]]
listings.head()
```

```python
#Convert price to flaot 
listings['Price']= listings['Price'].astype(float)
listings.head()
```

```python
listings = listings[listings.Odometer !="N/A"]
listings.head()
```

```python
# Remove comma in Odometer so you can convert to a float
listings['Odometer'] = listings['Odometer'].str.replace(',','')
listings
```

```python
#Convert to a float 
listings['Odometer']= listings['Odometer'].astype(float)
listings.head()
```

```python
listings = listings.groupby(['Manufacturer','Odometer']).mean()
listings.Price = listings.Price.round()
listings.head(50)
```

```python
listings = listings.sort_values("Odometer", ascending = False)
listings
```

```python
# Set as table
listings.to_json(orient='table')
```

```python
# Convert listings dataframe to json & name new json file
listings.to_json(r'depreciation_age_milage.json', orient='table')
```
